```python
# Eliminar una celda o valor específico.
## Primero debes acceder con 'at', 'loc' o 'iloc'

import numpy as np
#Puedes usar tambien pd.NA , Null ,''
df_2.at[3,'nombre'] = np.nan
df_2
```

```python
# df_2 = df_2[(df_2.iloc[:,0] =='Barranquilla')]
# df_2 = df_2[(df_2.loc[:,'ciudad'] =='Barranquilla')]
df_filtro = ((df_2['ciudad'] == 'Barranquilla') & ( pd.isna(df_2['nombre'])))
#Aquí obtenemos lista de T/F
df_filtro_index = df_2.index[df_filtro]
#Aquí obtenemos el indice que cumple con las consiciones
df_2 = df_2.drop(df_filtro_index)
df_2 = df_2.reset_index(drop=True)
df_2
```

```python
#Obtener lista de True/False, sirve dentro del drop

df_filtro1 = df_2['ciudad'] == 'Barranquilla'
#Obtenemos lista de true/false
df_filtro1_index = df_2.index[df_filtro1]
#Obtenemos lista de indices
df_filtro2 = df_2[df_filtro1]
df_filtro2 = df_filtro2.reset_index(drop=True)
df_filtro2
#df_filtro2
```

## Conclusión de practica inmediatamente anterior:
###### al aplicar condicionales con los df se obtienen listas de T/F
###### al aplicar df.index pasandole una lista de T/F
###### se obtiene un array de los indices que cumplen.
###### al pasarle a drop los indices anteriores, los procederá a eliminar.
###### Si con drop quiero eliminar una columna debo especificar 
###### el nombre así: columns = 'nombre_de_la_columna'
###### Si quiero obtener los valores que solo cumplen
###### se procede a ingresar la lista de T/F en el df[lista T/F]
